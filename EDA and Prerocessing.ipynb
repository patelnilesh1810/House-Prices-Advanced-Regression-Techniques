# EDA and Preprocessing

```python
import numpy as np
import pandas as pd
pd.options.display.max_columns = 100
```

```python
#Reading train data
train_data  = pd.read_csv("train.csv")


#Reading test data
test_data = pd.read_csv('test.csv')
```

```python
#contatinate train and test data for Preprocessing

total_data = pd.concat([train_data, test_data], ignore_index=True)
```

```python
total_data.shape
```

```python
total_data.tail()
```

```python
#dropping unwanted columns

total_data.drop(['LotFrontage', 'Street', 'Alley', 'Utilities',
                 'YearRemodAdd','YrSold',  'MiscFeature','Functional'],axis = 1,inplace = True)
```

```python
total_data.shape
```

```python
#Creating 'YearBuilt_age' column by subtracting 'yearBuilt' from 2018
#dropping 'yearBuilt' column

total_data['YearBuilt_age'] = 2018 - total_data['YearBuilt']
total_data.drop('YearBuilt',axis = 1, inplace=True)
```

```python
# Dropping 'BsmtFinSF1','BsmtFinSF2','BsmtUnfSF','1stFlrSF','2ndFlrSF' columns
#these columns information is included in other two columns in data

total_data.drop(['BsmtFinSF1','BsmtFinSF2','BsmtUnfSF','1stFlrSF','2ndFlrSF'],axis = 1, inplace=True)
```

```python
#Creating a new column "TotalLivingArea" by adding these columns

total_data['TotalLivingArea'] = total_data[['MasVnrArea', 'TotalBsmtSF',  'LowQualFinSF', 'GrLivArea', 
                                            'GarageArea', 'WoodDeckSF', 'OpenPorchSF', 'EnclosedPorch', 
                                            '3SsnPorch', 'ScreenPorch', 'PoolArea']].sum(axis=1)
```

```python
#dropping columns

total_data.drop(['MasVnrArea', 'TotalBsmtSF',  'LowQualFinSF', 'GrLivArea',
                 'GarageArea', 'WoodDeckSF','OpenPorchSF', 'EnclosedPorch',
                 '3SsnPorch', 'ScreenPorch', 'PoolArea'],axis = 1, inplace = True)
```

```python
#Changing ratings(categorical values) by numbers
Rating = {'Ex': 9, 'Gd':7, 'TA': 5, 'Fa':3, 'Po': 2,'NA':0}
```

```python
total_data['ExterQual'] = total_data['ExterQual'].map(Rating)
total_data['ExterCond'] = total_data['ExterCond'].map(Rating)
total_data['BsmtCond'] = total_data['BsmtCond'].map(Rating)
total_data['BsmtQual'] = total_data['BsmtQual'].map(Rating)
total_data['GarageCond'] = total_data['GarageCond'].map(Rating)
total_data['GarageQual'] = total_data['GarageQual'].map(Rating)
total_data['HeatingQC'] = total_data['HeatingQC'].map(Rating)
total_data['KitchenQual'] = total_data['KitchenQual'].map(Rating)
total_data['FireplaceQu'] = total_data['FireplaceQu'].map(Rating)
total_data['PoolQC'] = total_data['PoolQC'].map(Rating)
```

```python
#Changing ratings(categorical values) by numbers
Rating1 = {'Gd':4, 'Av': 3, 'Mn':2, 'No': 1,'NA':0}
total_data['BsmtExposure'] = total_data['BsmtExposure'].map(Rating1)
```

```python
#Creating a new column 'NonKitchenBedroomAbvGr' by other columns
total_data['NonKitchenBedroomAbvGr'] = (total_data['TotRmsAbvGrd']) - (total_data['BedroomAbvGr'] + total_data['KitchenAbvGr'])
```

```python
##Creating 'GarageYrBlt_age' column by subtracting 'GarageYrBlt_age' from 2018
total_data['GarageYrBlt_age'] = 2018 - total_data['GarageYrBlt']
```

```python
#Creating total Bathrooms column
#by multiply halfbathrooms with 0.5 and adding it to full bathroom

total_data['HalfBath'] = .5 * total_data['HalfBath']
total_data['BsmtHalfBath'] = .5 * total_data['BsmtHalfBath']

total_data['BathRooms'] = train_data[['BsmtFullBath','BsmtHalfBath','FullBath','HalfBath']].sum(axis=1)
```

```python
#dropping unwanted columns
total_data.drop(['BsmtFullBath','BsmtHalfBath','FullBath','HalfBath','MiscVal','TotRmsAbvGrd',
                 'GarageYrBlt','Fence','PoolQC','Condition2','Exterior2nd'],axis = 1,inplace = True)
```

```python
#filing Na with 0 because in that places there is no garage or pool or basement or Fireplace

total_data.GarageQual = total_data.GarageQual.fillna(0)
total_data.GarageCond = total_data.GarageCond.fillna(0)
total_data.BsmtCond = total_data.BsmtCond.fillna(0)
total_data.BsmtExposure = total_data.BsmtExposure.fillna(0)
total_data.BsmtQual = total_data.BsmtQual.fillna(0)
total_data.FireplaceQu = total_data.FireplaceQu.fillna(0)

total_data.BsmtFinType1 = total_data.BsmtFinType1.fillna('No')
total_data.BsmtFinType2 = total_data.BsmtFinType2.fillna('No')


total_data.GarageType = total_data.GarageType.fillna('No')
total_data.GarageFinish = total_data.GarageFinish.fillna('No')

#filling Na with majority

total_data['Electrical'] = total_data['Electrical'].fillna('SBrkr')
total_data['Exterior1st'] = total_data['Exterior1st'].fillna('VinylSd')
total_data.GarageCars = total_data.GarageCars.fillna(2)
total_data.MSZoning = total_data.MSZoning.fillna('RL')
total_data.KitchenQual = total_data.KitchenQual.fillna(5)
total_data.MasVnrType = total_data.MasVnrType.fillna('None')
total_data.SaleType = total_data.SaleType.fillna('WD')
```

```python
#assigning ranking to categorical values
dic = {'GLQ':6, 'ALQ':5, 'Unf':1, 'Rec':3, 'BLQ':4, 'No':0, 'LwQ':2}
total_data['BsmtFinType1'] = total_data['BsmtFinType1'].map(dic)
total_data['BsmtFinType2'] = total_data['BsmtFinType2'].map(dic)
```

```python
#assigning numerical values to categorical scale values
dic1 = {'RFn':2, 'Unf':1, 'Fin':3, 'No':0}
total_data['GarageFinish'] = total_data['GarageFinish'].map(dic1)
```

```python
data_dummified = pd.get_dummies(columns= ['CentralAir','Condition1','Electrical','Exterior1st','Foundation','GarageType',
                      'Heating','HouseStyle','LandContour','LandSlope','LotConfig','MSZoning','MasVnrType',
                      'PavedDrive','RoofMatl','SaleCondition','SaleType','RoofStyle'],
                         data = total_data,drop_first=True)
```

```python
data_dummified.shape
```

```python
data_dummified.MSSubClass = data_dummified['MSSubClass'].astype(str)
```

```python
data_dummified = pd.get_dummies(columns= ['BldgType','LotShape','MSSubClass','Neighborhood'],data = data_dummified,drop_first=True)
```
